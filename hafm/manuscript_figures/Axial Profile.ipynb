```python
from matplotlib import pyplot
import numpy as np
import pandas as pd
```

```python
# Set the font family and size to use for Matplotlib figures.
pyplot.rcParams['font.family'] = 'serif'
pyplot.rcParams['font.size'] = 20
```

```python
data_Vel = pd.read_csv('troubleshoot/HTC6/centerline_vel.csv')
data_Temp = pd.read_csv('troubleshoot/HTC6/centerline_temp.csv')
```

```python
zv = data_Vel['Z [ m ]']
vel = data_Vel['Velocity [ m s^-1 ]']
zt = data_Temp['Z [ m ]']
temp = data_Temp['Temperature [ K ]']
```

```python
pyplot.figure(figsize=(14,8))
pyplot.title('Axial Velocity (x,y=0)')
pyplot.xlabel('Z [mm]')
pyplot.ylabel('Velocity [m/s]')
pyplot.plot(zv,vel,'o-',color='black', linewidth=4)
# pyplot.plot(zt,temp)
pyplot.vlines(0,0,1.1,color='lightgray', label='Inlet')
pyplot.vlines(0.00217435,0,1.1,color='red', label='Coils')
pyplot.vlines(0.00906664,0,1.1,color='red')
pyplot.fill_between([0.00217435,0.00906664],[1.1,1.1],color='red',alpha=0.5)
pyplot.vlines(0.0370883,0,1.1,color='blue', label='Mouthpiece')
pyplot.vlines(0.0510359,0,1.1,color='blue')
pyplot.fill_between([0.0370883,0.0510359],[1.1,1.1],color='blue',alpha=0.5)
# pyplot.vlines(   ,0,1.1,color='green', label='OpenAir Start')
pyplot.vlines(0.0451281,0,1.1,color='limegreen',label='Open Air',alpha=0.5)
pyplot.vlines(0.06198888,0,1.1,color='limegreen')
pyplot.fill_between([0.0451281,0.06198888],[1.1,1.1],color='limegreen',alpha=0.5)
pyplot.fill_between([0,0.00217435],[1.1,1.1],color='lightgray',alpha=0.5,)
pyplot.fill_between([0.00906664,0.0370883],[1.1,1.1],color='darkgray',label='Pipe')
pyplot.legend()
pyplot.xlim([-0.005,0.1])
pyplot.savefig('troubleshoot/axialVelocity.png', dpi=300) 
```

```python
pyplot.figure(figsize=(14,8))
pyplot.title('Axial Temperature (x,y=0)')
pyplot.xlabel('Z [mm]')
pyplot.ylabel('Velocity [m/s]')
pyplot.plot(zt,temp,'o-',color='black', linewidth=4,markersize=6)
# pyplot.plot(zt,temp)
pyplot.vlines(0,0,350,color='lightgray', label='Inlet')
pyplot.vlines(0.00217435,0,350,color='red', label='Coils')
pyplot.vlines(0.00906664,0,350,color='red')
pyplot.fill_between([0.00217435,0.00906664],[350,350],color='red',alpha=0.5)
pyplot.vlines(0.0370883,0,350,color='blue', label='Mouthpiece')
pyplot.vlines(0.0510359,0,350,color='blue')
pyplot.fill_between([0.0370883,0.0510359],[350,350],color='blue',alpha=0.5)
# pyplot.vlines(   ,0,1.1,color='green', label='OpenAir Start')
pyplot.vlines(0.0451281,0,350,color='limegreen',label='Open Air',alpha=0.5)
pyplot.vlines(0.06198888,0,350,color='limegreen')
pyplot.fill_between([0.0451281,0.06198888],[350,350],color='limegreen',alpha=0.5)
pyplot.fill_between([0,0.00217435],[350,350],color='lightgray',alpha=0.5,)
pyplot.fill_between([0.00906664,0.0370883],[350,350],color='darkgray',label='Pipe')
pyplot.legend()
pyplot.xlim([-0.005,0.1])
pyplot.ylim([290,335])
pyplot.savefig('troubleshoot/axialTemp.png', dpi=300) 
```
