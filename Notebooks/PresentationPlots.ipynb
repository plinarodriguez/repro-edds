```python
from matplotlib import pyplot
import numpy as np
import pandas as pd
```

```python
# Set the font family and size to use for Matplotlib figures.
pyplot.rcParams['font.family'] = 'serif'
pyplot.rcParams['font.size'] = 20
```

```python
x = np.arange(0,5,0.1)
y1 = 15*x
y2 = 50*x
y3 = 100*x
```

```python
pyplot.figure(figsize=(6,4))
# pyplot.title('Model vs. Compute Time')
pyplot.xlabel('Mesh Size')
pyplot.ylabel('Compute Time')
pyplot.plot(x[20:],y1[20:],label='Machine 1',linewidth=4)
pyplot.plot(x[20:],y2[20:],label='Machine 2',linewidth=4)
pyplot.plot(x[20:],y3[20:],label='Machine 3',linewidth=4)
pyplot.legend(loc='upper left')
pyplot.xticks([])
pyplot.yticks([])
pyplot.ylim([0,700])
# pyplot.xlim([2,])
pyplot.savefig('machineMeshTime.png', dpi=300) 
```

```python
import random

i = 0 
pyplot.figure(figsize=(6,4))
# pyplot.title('CM&S - 20 Samples')
pyplot.xlabel('time')
pyplot.ylabel('QOI')
ysamples = []
while i < 15:
    a = random.uniform(0.5,1.5)
    y = a*x
    ysamples.append(y)
    pyplot.plot(x,y)
    i+=1
# pyplot.plot(x,x*1.5)
pyplot.plot(x,x, color='k',linestyle='dashed',linewidth=3,label='Nominal')
pyplot.yticks([])
pyplot.xticks([])
pyplot.legend()
pyplot.savefig('simSamples.png', dpi=300) 
```

```python
t = 0
errors = []
errors2 = []
meansamples = []
while t < len(x):
    temp = []
    for ys in ysamples:
        temp.append(ys[t])
    meansamples.append(np.average(temp))
    errors.append(np.std(temp))
    errors2.append(2*np.std(temp))
    t+=1
maxerror = np.max(errors)
```

```python
i = 0 
pyplot.figure(figsize=(6,4))
# pyplot.title('CM&S - 20 Samples')
pyplot.xlabel('time')
pyplot.ylabel('QOI')
pyplot.fill_between(np.array(x),np.array(x)+errors2,np.array(x)-errors2, color="lightgray",label='SD')
pyplot.plot(x,x,color='k',linewidth=3,label='Mean')
pyplot.yticks([])
pyplot.xticks([])
pyplot.legend(loc='upper left')
pyplot.savefig('simSamplesMeanSD.png', dpi=300) 
```

```python
data_Vel = pd.read_csv('troubleshoot/HTC6/centerline_vel.csv')
data_Temp = pd.read_csv('troubleshoot/HTC6/centerline_temp.csv')
```

```python
zv = data_Vel['Z [ m ]']
vel = data_Vel['Velocity [ m s^-1 ]']
zt = data_Temp['Z [ m ]']
temp = data_Temp['Temperature [ K ]']
```
